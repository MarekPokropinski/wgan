```python
import os
os.environ["CUDA_VISIBLE_DEVICES"] = "-1"

import tensorflow as tf
import tqdm
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt

from wgan import WGAN
from models.good_models import Discriminator, Generator
from models.mnist_models import MnistDiscriminator, MnistGenerator
```

```python
z_dim=128

generator = Generator(out_channels=3, dim=128)
generator.build((1, 128))
generator.load_weights('saved_models/faces/generator.keras')
```

```python
samples = 10
z = tf.random.normal((samples, z_dim))
pred = generator.predict(z)
```

```python
samples = 36
z = tf.random.normal((samples, z_dim))
pred = generator.predict(z)

fig, ax = plt.subplots(6, 6)
for i in range(samples):
    ax[i%6, i//6].imshow((pred[i]+1.0)/2.0)
    ax[i%6, i//6].axis('off')
```
